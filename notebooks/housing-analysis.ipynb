# Housing data prediction
Goal is to predict either the price of a house or area (size).

```python
import pandas as pd
import matplotlib
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/housing.csv')
```

```python
df
```

## Numerical boolean denotation
0 = False
1 = True

## furnishingstatus denotation
0 = Unfurnished
1 = Semi-furnished
2 = Furnished

```python
for col in df.columns:
    if isinstance(df[col][0], str):
        if col != 'furnishingstatus':
            df[col].replace({'no': 0, 'yes': 1}, inplace=True)
        else:
            df[col].replace({'unfurnished': 0, 'semi-furnished': 1, 'furnished': 2}, inplace=True)
```

## Data viz to determine appropriate model
